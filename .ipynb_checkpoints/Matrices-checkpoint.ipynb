 A matrix is a collection of values stored as rows and columns
 
 
 
 
 

```python
import numpy as np
```

In Numpy, there are two main ways to work with matrices - 

1. the np.array() method 
2. the np.matrix() method. Although the np.matrix() method may look more convenient at first, it only allows you to work with two dimensional matrices (a major drawback). 

In most cases, you will need to work with 3d, 4d etc. matrices, which you can do with the np.array() method.

```python
A = np.array([[1, 2, 3],
             [5, 6, 7]])
print(A.shape)
```

```python
print(A[:,0]) #to get first column
```

```python
print(A[0,:]) #to get get row
```

### Matrix operations in Numpy

```python
A = np.array([[1, 2, 1], [2, 1 ,2]]) # 2 by 3 matrix

B = np.array([[5, 2, 1], [2, 8 ,2]]) 

C= np.array([[5, 2], [8 ,2], [5, 9]]) #3 by 2 matrix 

print("Addition : \n" , A+B)
```

```python
#Multiplication
print("Multiplication : \n", np.dot(A,C))
```

```python
#Transpose
print("Transpose: \n", A.T)
```

### Inverse

```python
#Inverse
A = np.array([[1, 2],
             [5, 7]])

print("Inverse of A: \n", np.linalg.inv(A))
```

```python
A = np.array([[1, 2, 3],
             [5, 7, 8],
             [6, 8, 7]])
```

```python
print("Inverse of A: \n", np.linalg.inv(A))
```

### Determinant

```python
A = np.array([[1, 2],
             [5, 7]])

print(np.linalg.det(A))
```

### System of Linear Equations

Ax = B <br>
then x= A(inv) . B

```python
A = np.array([[1, 5, -1], 
              [2, 3, -2], 
              [-3, 4, 0]])

b = np.array([1, 2, 0])

A_inv = np.linalg.inv(A) #invers of A

x = np.dot(A_inv, b) # Calculate the value of parameter satisfying the equations
 
print(x)
```
