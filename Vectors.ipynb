The dot product can be used to compute the angle between two vectors.

Importance in ML is dot product of two vectors<br>

```python
# dot product using np

import numpy as np

a = np.array([2, 3, 5])
b = np.array([5, 6, 8])

print( "a.b = ", np.dot(a,b))
```

```python
# Calculate the angle between two vectors
import math
pi = math.pi

a = np.array([2, 3, 5])
b = np.array([5, 6, 8])

# a.b = mod(a) * mod(b) * cos(theta)

mod_a = np.linalg.norm(a)
print("Mod a ",mod_a)

mod_b = np.linalg.norm(b)
print("Mod b ", mod_b)

dot_ab= np.dot(a,b)
print("Dot product of a b ",dot_ab)

# To calculate theta
theta = np.arccos(dot_ab/(mod_a * mod_b))*(180/pi)
print("Theta is: ", theta)
```

## Vector Spaces

- Linear combination of vectors
- Basis and span 
- Vector spaces

1. The vectors ^i and ^j are chosen as the default basis vectors, though you can choose to have a completely different, valid choice of basis vectors (we'll discuss basis vectors and these choices in detail shortly)

2. The linear combination of two vectors is the sum of the scaled vectors 
3. The span of two or more vectors is the set of all possible vectors that you can get by changing the scalars and adding them
4. A set of vectors is called linearly dependent if any one or more of the vectors can be expressed as a linear combination of the other vectors. Geometrically, this means that if you remove some vectors from the set, you will still have the same span - some vectors are simply not adding anything to the span.
5. On the other hand, if none of the vectors in a set can be expressed as a linear combination of the other vectors, the vectors are called linearly independent.
